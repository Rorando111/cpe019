### **Importing libraries and dataset to be used**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, precision_recall_curve, roc_auc_score, roc_curve, accuracy_score
from sklearn.ensemble import RandomForestClassifier

import seaborn as sns

%matplotlib inline
```

```python
from keras.models  import Sequential
from keras.layers import Input, Dense, Flatten, Dropout, BatchNormalization
from keras.optimizers import Adam, SGD, RMSprop
```

```python
data = pd.read_csv('/content/apple_quality.csv')
data.head()
```

```python
data.info()
```

### Cleaning and processing data

- For this dataset, I had to remove unnecessary columns and change the target variable to binary so that it can fit into my model.

```python
data.dropna(inplace = True)
```

```python
data.drop(columns = ['A_id'], inplace = True)
```

```python
data.info()
```

```python
data["Quality"] = data["Quality"].apply(lambda toLabel: 0 if toLabel == 'bad' else 1)
```

```python
data
```

### **Splitting train and test samples**

```python
X = data.iloc[:,:-1].values
Y = data.iloc[:,-1].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, random_state = 1111, test_size = 0.25)
```

### **Scaling data**

```python
normalizer = StandardScaler()
X_train_norm = normalizer.fit_transform(x_train)
X_test_norm = normalizer.transform(x_test)
```

### **Creating base model**

```python
model_1 = Sequential()
model_1.add(Dense(12, input_shape=(7,), activation="relu"))
model_1.add(Dense(12, activation="relu"))
model_1.add(Dense(1, activation="sigmoid"))

model_1.compile(optimizer=SGD(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
run_hist_1 = model_1.fit(X_train_norm, y_train, validation_data=(X_test_norm, y_test), epochs=1000)

avg_accuracy = np.mean(run_hist_1.history['accuracy'])
print("Average Accuracy: {}%".format((avg_accuracy).round(3)*100))

avg_val_accuracy = np.mean(run_hist_1.history['val_accuracy'])
print("Average Validation Accuracy: {}%".format((avg_val_accuracy).round(3)*100))
```

Observation:
- The base model for this dataset resulted in 86.9% accuracy and 84.7% validation accuracy. Based on this results, we could say that the model is training well. However, this model can be further improved.

### **K-fold Cross Validation of base model**

```python
import tensorflow
```

```python
from sklearn.model_selection import KFold
from tensorflow.keras.callbacks import EarlyStopping

k = 5
kf = KFold(n_splits=k, random_state=None, shuffle=True)

fold_accuracies = []
train_loss = []
train_acc = []
val_loss = []
val_acc = []

for train_index, val_index in kf.split(X_train_norm):
    X_fold_train, X_fold_val = X_train_norm[train_index], X_train_norm[val_index]
    y_fold_train, y_fold_val = y_train[train_index], y_train[val_index]
    #Early stopping
    early_stopping = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)
    #Train model
    run_hist_fold = model_1.fit(X_fold_train, y_fold_train, validation_data=(X_fold_val, y_fold_val), epochs=1000, callbacks=[early_stopping])
    if early_stopping.stopped_epoch:
        print(f"Early stopping at epoch {early_stopping.stopped_epoch}")
    fold_accuracy = run_hist_fold.history['accuracy'][-1]
    train_loss.append(run_hist_fold.history['loss'][0])
    train_acc.append(run_hist_fold.history['accuracy'][0])
    val_loss.append(run_hist_fold.history['val_loss'][-1])
    val_acc.append(run_hist_fold.history['val_accuracy'][-1])
    fold_accuracies.append(fold_accuracy)

#Print the average accuracies of results
print("Mean fold accuracy: {}%".format(np.mean(fold_accuracies).round(3)*100))
print("Mean train loss: {}%".format(np.mean(train_loss).round(3)*100))
print("Mean train accuracy: {}%".format(np.mean(np.mean(train_acc)).round(3)*100))
print("Mean val loss: {}%".format(np.mean(np.mean(val_loss)).round(3)*100))
print("Mean val accuracy: {}%".format(np.mean(np.mean(val_acc)).round(3)*100))
```

**Analysis:**
- I used 5 folds for the cross validation, with an average fold accuracy of 92.2%. An average of 92.2% for the performance evaluation of this model can be further improved. In addidtion, I decided to use the early stopping function because I have observed from my earlier testing that the validation was overfitting.

### **Improving Accuracy**

```python
import tensorflow
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# Define the model
model_2 = Sequential()
model_2.add(Dense(20, input_shape=(7,), activation='relu'))
model_2.add(Dense(20, activation='relu'))
model_2.add(Dense(20, activation='relu'))
model_2.add(Dense(1, activation='sigmoid'))

early_stopping = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)

model_2.compile(optimizer=SGD(learning_rate=0.03), loss='binary_crossentropy', metrics=['accuracy'])

run_hist_2 = model_2.fit(X_train_norm, y_train, validation_data=(X_test_norm, y_test), epochs=1500, callbacks=[early_stopping])

avg_accuracy = np.mean(run_hist_2.history['accuracy'])
print("Average Accuracy: {}%".format((avg_accuracy).round(3)*100))

avg_val_accuracy = np.mean(run_hist_2.history['val_accuracy'])
print("Average Validation Accuracy: {}%".format((avg_val_accuracy).round(3)*100))
```

**Analysis:**
- For the modification of the base model, I decided to add another hidden layer with 20 nodes and an activation of 'relu'. I also decided to use early stopping for this model in order to  prevent overfitting, but it seems like there wasn't any need. Observing the accuracy results, with an average accuracy of 92% and average validation accuracy of 89.9%, this model clearly shows better performance compared to the base model.

```python
from sklearn.model_selection import KFold
from tensorflow.keras.callbacks import EarlyStopping

k = 5
kf = KFold(n_splits=k, random_state=None, shuffle=True)


fold_accuracies = []
train_loss = []
train_acc = []
val_loss = []
val_acc = []

for train_index, val_index in kf.split(X_train_norm):
    X_fold_train, X_fold_val = X_train_norm[train_index], X_train_norm[val_index]
    y_fold_train, y_fold_val = y_train[train_index], y_train[val_index]

    early_stopping = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)

    run_hist_fold = model_2.fit(X_fold_train, y_fold_train, validation_data=(X_fold_val, y_fold_val), epochs=1500, callbacks=[early_stopping])
    if early_stopping.stopped_epoch:
        print(f"Early stopping at epoch {early_stopping.stopped_epoch}")

    fold_accuracy = run_hist_fold.history['accuracy'][-1]
    train_loss.append(run_hist_fold.history['loss'][0])
    train_acc.append(run_hist_fold.history['accuracy'][0])
    val_loss.append(run_hist_fold.history['val_loss'][-1])
    val_acc.append(run_hist_fold.history['val_accuracy'][-1])
    fold_accuracies.append(fold_accuracy)



print("Mean fold accuracy: {}%".format(np.mean(fold_accuracies).round(3)*100))
print("Mean train loss: {}%".format(np.mean(train_loss).round(3)*100))
print("Mean train accuracy: {}%".format(np.mean(np.mean(train_acc)).round(3)*100))
print("Mean val loss: {}%".format(np.mean(np.mean(val_loss)).round(3)*100))
print("Mean val accuracy: {}%".format(np.mean(np.mean(val_acc)).round(3)*100))
```

**Analysis:**
- The k-fold cross validation also concludes that this model performs better and has a high accuracy.

## **Regression**

### **Importing libraries and dataset to be used**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, precision_recall_curve, roc_auc_score, roc_curve, accuracy_score
from sklearn.ensemble import RandomForestClassifier

import seaborn as sns

%matplotlib inline
```

```python
from keras.models  import Sequential
from keras.layers import Input, Dense, Flatten, Dropout, BatchNormalization
from keras.optimizers import Adam, SGD, RMSprop
```

```python
df = pd.read_csv('/content/Student_Performance.csv')
```

### **Cleaning and processing data**

```python
df.head()
```

```python
df.info()
```

```python
df["Extracurricular Activities"] = df["Extracurricular Activities"].apply(lambda toLabel: 0 if toLabel == 'No' else 1)
```

```python
df.head()
```

### **Splitting train and test samples**

```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, random_state = 1111, test_size = 0.25)
```

### **Creating base model**

```python
model = Sequential()
model.add(Dense(20, input_shape=(5,), activation='relu'))
model.add(Dense(1))

model.compile(optimizer="rmsprop", loss="mse", metrics=["mae"])
hist_3 = model.fit(X_train, Y_train, epochs=100, batch_size=32, validation_data=(X_test, Y_test))

# Print the MSE for the training and testing sets
print("Base model MSE:")
print("Average Training MSE: ", hist_3.history['loss'][-1])
print("Average Validation MAE: ", hist_3.history['val_loss'][-1])
```

**Analysis:**
- For the base model's architecture, it only consists of 1 hidden layer and 20 nodes. The training process was for 100 epochs, with a batch size of 32, using 'rmsprop' as optimizer and mse as loss function. While the testing MSE fluctuated but generally showed a declining trend, the training MSE showed a consistent decline throughout the course of the epochs. The Average training MSE was 4.628, and the Average Valdidation MAE of 4.825. This implies that although the model fits the data rather well, it could be a little overfitting. It could take further assessment to improve the model's functionality.

### **Scaling data**

```python
normalizer = StandardScaler()
X_train_scaled = normalizer.fit_transform(X_train)
X_test_scaled = normalizer.transform(X_test)
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model_1 = Sequential()
model_1.add(Dense(12, input_shape=(5,), activation="relu"))
model_1.add(Dense(12, activation="relu"))
model_1.add(Dense(1, activation="linear"))

model_1.compile(optimizer="rmsprop", loss="mse", metrics=["mae"])
# Train the model
hist_4 = model_1.fit(X_train_scaled, Y_train, epochs=100, batch_size=32, validation_data=(X_test_scaled, Y_test))
# Print the MSE for the training and testing sets
print("Scaled model MSE:")
print("Average Training MSE: ", hist_4.history['loss'][-1])
print("Average Validation MAE: ", hist_4.history['val_loss'][-1])
```

### **Tuning using Larger model**

```python
# Try a larger model
model_2 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(24, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    tf.keras.layers.Dense(24, activation='relu'),
    tf.keras.layers.Dense(24, activation='relu'),
    tf.keras.layers.Dense(16, activation='linear'),
    tf.keras.layers.Dense(1)
])

# Compile the model
model_2.compile(optimizer="adam", loss="mse", metrics=["mse"])

# Train the model
hist_L = model_2.fit(X_train_scaled, Y_train, epochs=100, batch_size=32, validation_data=(X_test_scaled, Y_test))

# Print the MSE for the training and testing sets
print("Larger model MSE:")
print("Training: ", hist_L.history['loss'][-1])
print("Testing: ", hist_L.history['val_loss'][-1])
```

### **Tuning using Smaller model**

```python
# Try a smaller model
model_3 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(8, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(1, activation='linear')
])

# Compile the model
model_3.compile(optimizer='adam', loss='mse')

# Train the model
hist_S = model_3.fit(X_train_scaled, Y_train, epochs=100, batch_size=32, validation_data=(X_test_scaled, Y_test))

print("Smaller model MSE:")
print("Training: ", hist_S.history['loss'][-1])
print("Testing: ", hist_S.history['val_loss'][-1])
```
